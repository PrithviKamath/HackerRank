```python
def minion_game(string):
    # your code goes here
    vowels=['A','E','I','O','U']
    Stuart={}
    Kevin={}
    for i in range(0,len(string)):
        if string[i] in vowels:
            #Kevin count
            for j in range(1,len(string)+1):
                if i+j <= len(string): #check if i+j is les than length of complete string
                    if not string[i:i+j] in Kevin:
                        Kevin[string[i:i+j]]=1 #create new key value pair if new key found
                    else:
                        Kevin[string[i:i+j]] +=1 #add value to key if key already found
        else:
            #Stuart count
            for j in range(1,len(string)+1):
                if i+j <= len(string):
                    if not string[i:i+j] in Stuart:
                        Stuart[string[i:i+j]] =1
                    else:
                        Stuart[string[i:i+j]] +=1
    #compare and print highest count or print draw
    if sum(Kevin.values()) > sum(Stuart.values()):
        print("Kevin " + str(sum(Kevin.values())))
    elif sum(Stuart.values()) > sum(Kevin.values()):
        print("Stuart "+ str(sum(Stuart.values())))
    else:
        print("Draw")
```

```python
def minion_game(string):
    # your code goes here
    vowels=['A','E','I','O','U']
    Stuart=0
    Kevin=0
    for i in range(0,len(string)):
        if string[i] in vowels:
            #Kevin count
            for j in range(1,len(string)+1):
                if i+j <= len(string):
                    Kevin +=1
        else:
            #Stuart count
            for j in range(1,len(string)+1):
                if i+j <= len(string):
                    Stuart +=1
    if Kevin > Stuart:
        print("Kevin " + str(Kevin))
    elif Stuart > Kevin:
        print("Stuart "+ str(Stuart))
    else:
        print("Draw")
```
